```python
from PIL import Image, ImageDraw
from google.colab import files
from IPython.display import display
uploaded = files.upload()
filename = next(iter(uploaded))
input_image = Image.open(filename).convert("RGB")
input_pixels = input_image.load()
box_kernel = [[1/9, 1/9, 1/9],
              [1/9, 1/9, 1/9],
              [1/9, 1/9, 1/9]]
offset = len(box_kernel) // 2
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)
for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]
        for a in range(len(box_kernel)):
            for b in range(len(box_kernel)):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                acc[0] += pixel[0] * box_kernel[a][b]
                acc[1] += pixel[1] * box_kernel[a][b]
                acc[2] += pixel[2] * box_kernel[a][b]
        draw.point((x, y), (int(acc[0]), int(acc[1]), int(acc[2])))
output_image.save("blurred.png")
display(output_image)
```

```python
from google.colab import files
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
uploaded = files.upload()
filename = next(iter(uploaded))
gaussian_kernel = (1/256) * np.array([
    [1, 4, 6, 4, 1],
    [4,16,24,16,4],
    [6,24,36,24,6],
    [4,16,24,16,4],
    [1, 4, 6, 4, 1]
], dtype=np.float32)
jpg = cv2.imread(filename)
if jpg is None:
    print("Không load được ảnh.")
else:
    blurred = cv2.filter2D(jpg, -1, gaussian_kernel)
    cv2_imshow(blurred)
```

```python
from PIL import Image, ImageDraw
from google.colab import files
from IPython.display import display
uploaded = files.upload()
filename = next(iter(uploaded))
input_image = Image.open(filename).convert("L")
input_pixels = input_image.load()
kernels = {
    'kernel_a': [[0, -1,  0],
                 [-1, 4, -1],
                 [0, -1,  0]],
    'kernel_b': [[-1, -1, -1],
                 [-1, 8, -1],
                 [-1, -1, -1]]
}
offset = 1
for name, kernel in kernels.items():
    output_image = Image.new("L", input_image.size)
    draw = ImageDraw.Draw(output_image)
    for x in range(offset, input_image.width - offset):
        for y in range(offset, input_image.height - offset):
            acc = 0
            for a in range(3):
                for b in range(3):
                    xn = x + a - offset
                    yn = y + b - offset
                    pixel = input_pixels[xn, yn]
                    acc += pixel * kernel[a][b]
            acc = max(0, min(255, int(acc)))
            draw.point((x, y), acc)
    output_image.save(f"{name}_edge.png")
    print(f"Ảnh lọc theo {name}:")
    display(output_image)
```
